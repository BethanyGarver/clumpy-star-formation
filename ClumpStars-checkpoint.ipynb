```python
from astropy.table import Table
import numpy as np
import pylab as plt
import photutils as phot
```

```python
def FitsRead(time):
    # Read the fits file for that time and return table of stars
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
txyn = Table.read('clumps5.fits',1)
stars = Table.read('clumps5.fits',2)
txyz = Table.read('txyzform.fits',1)
idlist = np.unique(np.array(txyn['CLUMPID']))
```

```python
for clumpid in idlist:
    masses = []
    times = []
    unbound = []
    txyclump = txyn[(txyn['CLUMPID']==clumpid)]
    starsclump = stars[(stars['CLUMPID']==clumpid)]
    starttime = int(txyclump['TIME'][0]*10)+1
    endtime = int(txyclump['TIME'][-1]*10)+2
    for time in range(starttime,endtime):
        #plt.figure(figsize=(10,10))
        snapshot = FitsRead(time)
        cutstars = starsclump[(starsclump['INDEX']<len(snapshot))]
        clumpstars = snapshot[cutstars['INDEX']]
        x = clumpstars['X']
        y = clumpstars['Y']
        tform = txyz['TFORM'][cutstars['INDEX']]
        #plt.scatter(x,y,c=tform,s=1)
        #plt.colorbar()
        try:
            current = txyclump[(txyclump['TIME'] == time/10)]
            clumpx = current['X']
            clumpy = current['Y']
            radius = 3*np.sqrt((sum((x-clumpx)**2)+sum((y-clumpy)**2))/len(x))
        except:
            clumpx = np.mean(x)
            clumpy = np.mean(y)
            radius = 3*np.sqrt((sum((x-clumpx)**2)+sum((y-clumpy)**2))/len(x))
        #aperture = phot.CircularAperture((clumpx,clumpy), r=radius)
        #aperture.plot(color='r', lw=1.5)
        #im, xedges, yedges  = np.histogram2d (-snapshot['Y'],snapshot['X'],(100,100),range=[[-7,7],[-7,7]])
        #extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
        #plt.imshow(np.log(im),cmap='Greys_r',extent=extent)
        #plt.xlabel('x')
        #plt.ylabel('y')
        bound = clumpstars[((clumpstars['X']-clumpx)**2+(clumpstars['Y']-clumpy)**2<radius**2)]
        mass = sum(bound['MS'])*2.33*10**5
        times.append(time/10)
        masses.append(mass)
        unbound.append(1-len(bound)/len(clumpstars))
    plt.figure(figsize=(20,6))
    plt.subplot(1,3,1)
    plt.scatter(times,masses)
    plt.xlabel("Time")
    plt.ylabel("Mass")
    plt.title("Bound mass of clump {}".format(clumpid))
    plt.subplot(1,3,2)
    plt.scatter(times,unbound)
    plt.xlabel('Time')
    plt.ylabel('Unbound fraction')
    plt.title('Fraction of unbound stars in clump {0}'.format(clumpid))
    plt.subplot(1,3,3)
    plt.scatter(clumpstars['FEH'],clumpstars['OFE'],c=tform,s=1)
    plt.colorbar()
    plt.xlim(-0.8,0.6)
    plt.ylim(-1,0.2)
    plt.xlabel('[Fe/H]')
    plt.ylabel('[O/Fe]')
    plt.title('Chemistry plot of clump {}'.format(clumpid))
```
