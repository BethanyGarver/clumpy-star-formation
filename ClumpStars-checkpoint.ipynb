```python
from astropy.table import Table
import numpy as np
import pylab as plt
import photutils as phot
```

```python
def FitsRead(time):
    # Read the fits file for that time and return table of stars
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
txy = Table.read('clumps3.fits',1)
stars = Table.read('clumps3.fits',2)
txyz = Table.read('txyzform.fits',1)
```

<h4>Table showing times clump 63 was tracked and the x and y positions of the clump at those times.</h4>

```python
txyin63 = (txy['CLUMPID']==63)
txy63 = txy[txyin63]
print(txy63)
```

<h4>Plot of the x and y positions of clump 63 as it spirals toward the center.</h4>

```python
x = txy63['X']
y = txy63['Y']
plt.figure(figsize=(10,10))
plt.plot(x,y)
```

<h4>Table showing indices (in the txyzform and snapshot files) for all the stars in clump 63.</h4>

```python
starsin63 = (stars['CLUMPID']==63)
stars63 = stars[starsin63]
print(stars63)
```

<h4>Plots of the x and y locations of the stars in clump 63 starting with the first snapshot after it started tracking the clump and ending with the first snapshot after the clump reaches the center. The color of the dots corresponds to their age at the time of the snapshot. The red circle has a radius of 0.5 and is centered on the (x,y) location of the clump at that time. The mass in the plot titles is the total mass of all the stars within a distance of 0.5 from the clump's center (assuming the clump's z position is 0).</h4>

```python
for time in range(20,28):
    plt.figure(figsize=(10,10))
    snapshot = FitsRead(time)
    cut = (stars63['INDEX']<len(snapshot))
    cutstars = stars63[cut]
    clumpstars = snapshot[cutstars['INDEX']]
    x = clumpstars['X']
    y = clumpstars['Y']
    age = time/10-txyz['TFORM'][cutstars['INDEX']]
    plt.scatter(x,y,c=age,s=1)
    plt.colorbar()
    current = (txy63['TIME'] == time/10)
    location = txy63[current]
    clumpx = location['X']
    clumpy = location['Y']
    aperture = phot.CircularAperture((clumpx,clumpy), r=0.5)
    aperture.plot(color='r', lw=1.5)
    im, xedges, yedges  = np.histogram2d (-snapshot['Y'],snapshot['X'],(100,100),range=[[-7,7],[-7,7]])
    extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
    plt.imshow(np.log(im),cmap='Greys_r',extent=extent)
    plt.xlabel('x')
    plt.ylabel('y')
    try:
        nearby = ((snapshot['X']-clumpx)**2+(snapshot['Y']-clumpy)**2+snapshot['Z']**2<0.25)
        stillthere = snapshot[nearby]
        mass = sum(stillthere['MS'])
        plt.title('Clump 63 stars at {0} billion years (color is age). Mass = {1}'.format(time/10,mass))
    except:
        plt.title('Clump 63 stars at {0} billion years (color is age).'.format(time/10))
```

<h4>Plots of the x and z positions of the stars in clump 63. Again the color corresponds to the age of the stars.</h4>

```python
for time in range(20,28):
    plt.figure(figsize=(20,10))
    snapshot = FitsRead(time)
    cut = (stars63['INDEX']<len(snapshot))
    cutstars = stars63[cut]
    clumpstars = snapshot[cutstars['INDEX']]
    x = clumpstars['X']
    z = clumpstars['Z']
    age = time/10-txyz['TFORM'][cutstars['INDEX']]
    plt.scatter(x,z,c=age,s=1)
    plt.colorbar()
    plt.xlabel('x')
    plt.ylabel('z')
    plt.title('Clump 63 stars at {0} billion years (color is age)'.format(time/10))
    im, xedges, yedges  = np.histogram2d (-snapshot['Z'],snapshot['X'],(100,100),range=[[-2,2],[-7,7]])
    extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
    plt.imshow(np.log(im),cmap='Greys_r',extent=extent)
```

<h4>Plots of [O/Fe] vs [Fe/H] at each snapshot for all the stars in clump 63 that had formed by the time of the snapshot.</h4>

```python
for time in range(20,28):
    plt.figure(figsize=(10,10))
    snapshot = FitsRead(time)
    cut = (stars63['INDEX']<len(snapshot))
    cutstars = stars63[cut]
    clumpstars = snapshot[cutstars['INDEX']]
    feh = clumpstars['FEH']
    ofe = clumpstars['OFE']
    age = time/10-txyz['TFORM'][cutstars['INDEX']]
    plt.scatter(feh,ofe,c=age,s=1)
    plt.colorbar()
    plt.xlabel('[Fe/H]')
    plt.ylabel('[O/Fe]')
    plt.xlim((-0.8,0.6))
    plt.ylim((-0.6,0.2))
    plt.title('[O/Fe] vs [Fe/H] for Clump 63 stars at {0} billion years (color is age)'.format(time/10))
```

<h4>Histogram showing the star formation rate in clump 63.</h4>

```python
plt.figure(figsize=(10,10))
time = txyz['TFORM'][stars63['INDEX']]
plt.hist(time, bins=250)
plt.xlabel('Time')
plt.ylabel('Stars formed in time step')
```

<h4>[O/Fe] vs [Fe/H] density plots for every 100 million years between 1.6 and 4 billion years. On the left are the plots for all the clump stars that had formed by that time and on the right are the plots for all stars that had formed.</h4>

```python
for time in range(16,41):
    snapshot = FitsRead(time)
    cut = (stars['INDEX']<len(snapshot))
    cutstars = stars[cut]
    clumpstars = snapshot[cutstars['INDEX']]
    plt.figure(figsize=(20,5))
    im, xedges, yedges  = np.histogram2d (-clumpstars['OFE'],clumpstars['FEH'],(100,100),range=[[-0.6,0.8],[-1.5,0.6]])
    extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
    plt.subplot(1,2,1)
    plt.gca().set_title('[O/Fe] vs [Fe/H] for clump stars at {0} billion years'.format(time/10))
    plt.imshow(im,extent=extent)
    plt.xlabel('[Fe/H]')
    plt.ylabel('[O/Fe]')
    plt.subplot(1,2,2)
    im, xedges, yedges  = np.histogram2d (-snapshot['OFE'],snapshot['FEH'],(100,100),range=[[-0.6,0.8],[-1.5,0.6]])
    extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
    plt.gca().set_title('[O/Fe] vs [Fe/H] for all stars at {0} billion years'.format(time/10))
    plt.imshow(im,extent=extent)
    plt.xlabel('[Fe/H]')
    plt.ylabel('[O/Fe]')
```
