```python
from astropy.table import Table
from astropy.table import vstack
import numpy as np
import pylab as plt
import photutils as phot
```

```python
def FitsRead(time):
    # Read the fits file for that time and return table of stars
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
txy = Table.read('clumps4.fits',1)
stars = Table.read('clumps4.fits',2)
merges = Table.read('clumps4.fits',3)
txyz = Table.read('txyzform.fits',1)
```

<h4>X-Y location plot and [O/Fe] vs [Fe/H] plot at 4 billion years for stars that formed within a distance of 0.3 from the center</h4>

```python
snapshot = FitsRead(40)
txyzcut = txyz[0:len(snapshot)]
center = ((txyzcut['XFORM']**2+txyzcut['YFORM']**2+txyzcut['ZFORM']**2)**(1/2)<0.3)
centerform = snapshot[center]
age = 4-txyzcut['TFORM'][center]
plt.figure(figsize=(10,10))
x = centerform['X']
y = centerform['Y']
plt.scatter(x,y,c=age,s=1)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y locations for stars formed in the center (color is age)')
plt.xlim((-25,25))
plt.ylim((-25,25))
plt.colorbar()
plt.figure(figsize=(10,10))
im, xedges, yedges  = np.histogram2d (-y,x,(100,100),range=[[-25,25],[-25,25]])
extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]
plt.imshow(np.log(im),cmap='Greys_r',extent=extent)
plt.xlabel('x')
plt.ylabel('y')
plt.title('x and y locations for stars formed in the center')
plt.xlim((-25,25))
plt.ylim((-25,25))
plt.figure(figsize=(10,10))
feh = centerform['FEH']
ofe = centerform['OFE']
plt.scatter(feh,ofe,c=age,s=1)
plt.xlabel('[Fe/H]')
plt.ylabel('[O/Fe]')
plt.title('[O/Fe] vs [Fe/H] for stars formed in the center (color is age)')
plt.xlim((-1.7,0.6))
plt.ylim((-0.8,0.6))
plt.colorbar()
```

<h4>IDs of clumps that spent any time in the center and IDs of clumps that spent more than 50 timesteps in the center</h4>

```python
inCenter = (txy['X']**2+txy['Y']**2<0.09)
center = txy[inCenter]

centerclumpslist = []
for i in center['CLUMPID']:
    if i not in centerclumpslist:
        centerclumpslist.append(i)
print(centerclumpslist)
print()

centerclumpslist2 = []
for i in centerclumpslist:
    inclumpi = (center['CLUMPID']==i)
    clumpi = center[inclumpi]
    if len(clumpi)>50:
        centerclumpslist2.append(i)
print(centerclumpslist2)
```

```python
tableslist = []
for i in centerclumpslist2:
    txyincenter = (txy['CLUMPID'] == i)
    center = txy[txyincenter]
    tableslist.append(center)
center = vstack(tableslist)

tableslist = []
for i in centerclumpslist2:
    starsincenter = (stars['CLUMPID'] == i)
    center = stars[starsincenter]
    tableslist.append(center)
centerstars = vstack(tableslist)
```

<h4>Plots of the x and y locations and [O/Fe] vs [Fe/H] of the stars in center clumps. The color of the dots corresponds to their age at the time of the snapshot.</h4>

```python
times = []
masses = []
for time in range(4,9):
    plt.figure(figsize=(20,10))
    snapshot = FitsRead(time*5)
    cut = (centerstars['INDEX']<len(snapshot))
    cutstars = centerstars[cut]
    clumpstars = snapshot[cutstars['INDEX']]
    plt.subplot(1,2,1)
    x = clumpstars['X']
    y = clumpstars['Y']
    age = time/2-txyz['TFORM'][cutstars['INDEX']]
    plt.scatter(x,y,c=age,s=1)
    plt.xlabel('x')
    plt.ylabel('y')
    plt.xlim((-10,10))
    plt.ylim((-10,10))
    plt.title('Center clump stars at {0} billion years (color is age).'.format(time/2))
    
    plt.subplot(1,2,2)
    feh = clumpstars['FEH']
    ofe = clumpstars['OFE']
    plt.scatter(feh,ofe,c=age,s=1)
    plt.xlabel('[Fe/H]')
    plt.ylabel('[O/Fe]')
    plt.xlim((-1,0.6))
    plt.ylim((-0.8,0.3))
    plt.title('[O/Fe] vs [Fe/H] for center clump stars at {0} billion years (color is age).'.format(time/2))    
    plt.colorbar()
```
