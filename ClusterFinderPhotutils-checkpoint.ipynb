```python
import photutils as phot
from astropy.stats import sigma_clipped_stats
from astropy.table import Table
import numpy as np
import scipy.stats as stats
from scipy.interpolate import UnivariateSpline
import pylab as plt
from astropy.visualization import simple_norm
from astropy.io import fits
```

```python
def FitsRead(time):
    # Read the fits file for that time and return table of stars
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
def RotateBack(x,y,dtime,vcirc):
    # Rotate stars back in time
    lindist = vcirc*dtime #linear distance traveled in time interval
    rho = (x**2+y**2)**(1/2) #distance from z axis
    nrot = lindist/(2*np.pi*rho) #number of rotations
    nrotints = []
    for i in nrot:
        nrotints.append(int(i))
    nrotints = np.array(nrotints) #integer number of rotations
    rotangle = nrot-nrotints #partial rotation
    rotanglerad = 2*np.pi*rotangle #convert to radians
    xrot = x*np.cos(-rotanglerad)-y*np.sin(-rotanglerad) #x position at beginning of interval
    yrot = x*np.sin(-rotanglerad)+y*np.cos(-rotanglerad) #y position at beginning of interval
    return xrot,yrot
```

```python
def make_unique(original_list):
    # Takes a list and returns it without duplicate entries
    unique_list = []
    [unique_list.append(obj) for obj in original_list if obj not in unique_list]
    return unique_list
```

```python
def StarsInInterval(starlist,start,end):
    # Returns table of stars that formed within time interval
    tRange = (starlist['TFORM']>start,starlist['TFORM']<end)
    inRange = np.array(np.zeros(len(tRange[0])),dtype=bool) 
    for i in range(0,len(inRange)):
        if tRange[0][i] and tRange[1][i]:
            inRange[i] = True            #true for stars that formed in time interval
    newStars = starlist[inRange]
    return newStars
```

```python
def FindPeaks(data,z):
    # Find peaks in 2D array of data
    statsdata = [] #only includes data points with density greater than 1 to calculate threshold
    for i in data:
        for j in i:
            if j>1:
                statsdata.append(j)
    mean, median, std = sigma_clipped_stats(statsdata, sigma=3.0)
    threshold = median + (z * std) #minimum density for a peak
    tbl = phot.find_peaks(data, threshold, box_size=11) #table of x index, y index, and density for peaks
    tbl['peak_value'].info.format = '%.8g'
    return tbl
```

```python
def CombineClose(positions):
    # Combine multiple peaks in the same star cluster
    peaks = []
    closelist = []
    for i in range(0,len(positions[0])):
        jlist = [i] #a list containing that peak and other peaks near it
        for j in range(i+1,len(positions[0])):
            xi = positions[0][i]
            yi = positions[1][i]
            xj = positions[0][j]
            yj = positions[1][j]
            dist2 = (xi-xj)**2+(yi-yj)**2
            if dist2<0.1:
                jlist.append(j)
        closelist.append(jlist) #a list where each element is a list of nearby peaks
    for i in range(0,len(closelist)):
        if len(closelist[i])>1:
            for j in closelist[i]:
                if i==j:
                    continue
                closelist[j] = closelist[0]
    closelist = make_unique(closelist) #remove elements of closelist that contain the same peaks
    for i in closelist:
        sumx = 0
        sumy = 0
        for j in i:
            sumx = sumx+positions[0][j]
            sumy = sumy+positions[1][j]
        meanx = sumx/len(i) #mean x position of a close group of peaks
        meany = sumy/len(i) #mean y position of a close group of peaks
        peaks.append([meanx,meany])
    xcoords = []
    ycoords = []
    for peak in peaks:
        xcoords.append(peak[0])
        ycoords.append(peak[1])
    xcoords = np.array(xcoords) #array of peak x coordinates
    ycoords = np.array(ycoords) #array of peak y coordinates
    peaks1 = (xcoords,ycoords)
    return peaks1
```

```python
def MeanPeaks(starlist,positions):
    # Make the x and y positions of the clumps equal to the mean x and y positions of the stars, rather than peak location
    x = []
    y = []
    for i in range(0,len(positions[0])):
        inclump = ((starlist['XROT']-positions[0][i])**2+(starlist['YROT']-positions[1][i])**2<0.09) #true for stars within
                                                                                                     #aperture
        clump = starlist[inclump] #only includes stars in clump
        x.append(np.mean(clump['XROT']))
        y.append(np.mean(clump['YROT']))
    x = np.array(x) #array of mean x coordinates of clump
    y = np.array(y) #array of mean y coordinates of clump
    meanPositions = (x,y)
    return x,y,meanPositions
```

```python
def TrackClumps(oldClumps,xcoords,ycoords,stars):
    # Match clumps in current timestep to clumps in previous timesteps
    leftx = xcoords
    lefty = ycoords
    lefti = []
    clumps = oldClumps
    for i in range(0,len(leftx)):
        lefti.append(i)
    if len(clumps)>0:
        for skip in range(1,6): #run through the previous 5 timesteps
            clumpsrho = (leftx**2+lefty**2)**(1/2)
            clumpsvcirc = vDisk(clumpsrho)
            cxrot,cyrot = RotateBack(leftx,lefty,skip*dt,clumpsvcirc) #move clumps back in time
            xleft = []
            yleft = []
            ileft = []
            for i in range(0,len(cxrot)): #run through clumps in current timestep
                for clump in clumps: #run through clumps from previous timesteps
                    newest = clump[len(clump)-1] #[time,x,y] of the most recent location of the clump
                    if newest[0]==round(start-skip*dt,3): #only look at clumps that were found recently
                        xi = cxrot[i]
                        yi = cyrot[i]
                        xj = newest[1]
                        yj = newest[2]
                        if (xi-xj)**2+(yi-yj)**2<0.4 and clumpsrho[i]>0.3:
                            inclump = ((stars['XROT']-leftx[i])**2+(stars['YROT']-lefty[i])**2<0.09)
                            clumpStars = stars[inclump] #only includes stars in clump
                            clump.append([start,xcoords[lefti[i]],ycoords[lefti[i]]])
                            clump[0] = clump[0]+list(clumpStars['INDEX'])
                            clump[0] = make_unique(clump[1])
                            break
                    if clump==clumps[len(clumps)-1]: #make a list of clumps that do not match clumps seen earlier
                        xleft.append(leftx[i])
                        yleft.append(lefty[i])
                        ileft.append(lefti[i])
            leftx = np.array(xleft)
            lefty = np.array(yleft)
            lefti = ileft
    for i in range(0,len(lefti)): #add clumps that don't match older ones to the end of the list
        if leftx[i]**2+lefty[i]**2>0.09:
            inclump = ((stars['XROT']-leftx[i])**2+(stars['YROT']-lefty[i])**2<0.09)
            clumpStars = stars[inclump] #only includes stars in clump        
            clumps.append([list(clumpStars['INDEX']),[start,xcoords[lefti[i]],ycoords[lefti[i]]]])
    return clumps
```

```python
# Read txyz file
txyz = Table.read('txyzform.fits',1)
index = np.linspace(0,len(txyz)-1,len(txyz))
txyz.add_column(Table.Column(index),name='INDEX')
rho = np.sqrt(txyz['XFORM']**2+txyz['YFORM']**2)
txyz.add_column(rho,name='RHO')
start = 1.5
dt = 0.001
clumps = []
oldClumps = []

for refTime in range(15,41):
    print('refTime = {}'.format(refTime))
    # Read snapshot file
    snapshot = FitsRead(refTime)
    r = np.sqrt(snapshot['X']**2+snapshot['Y']**2)
    theta = np.arctan2(snapshot['Y'],snapshot['X'])
    vth = -snapshot['VX']*np.sin(theta)+snapshot['VY']*np.cos(theta)
    snapshot.add_column(r,name='R')
    snapshot.add_column(vth,name='VTH')

    # Interpolate vDisk as a function of rho
    bins = int(np.max(r)/2.5)
    medians,binEdges,binNumber = stats.binned_statistic(r,vth,'median',bins)
    edges = []
    for i in range(0,len(binEdges)-1):
        edges.append((binEdges[i]+binEdges[i+1])/2)
    w = np.isnan(medians)
    medians[w] = 0
    vDisk = UnivariateSpline(edges,medians,~w)

    while round(start*10)==refTime and start<4:
        #Find stars that formed in time interval
        end = start+5*dt
        newStars = StarsInInterval(txyz,start,end)

        # Move stars back in time to beginning of interval
        vcirc = vDisk(newStars['RHO'])
        dtime = newStars['TFORM']-start
        xrot,yrot = RotateBack(newStars['XFORM'],newStars['YFORM'],dtime,vcirc)
        newStars.add_column(xrot,name='XROT')
        newStars.add_column(yrot,name='YROT')
        im, xedges, yedges  = np.histogram2d (-yrot,xrot,(50,50))
        dx = xedges[1]-xedges[0]
        dy = yedges[1]-yedges[0]
        extent = [yedges[0], yedges[-1], -xedges[-1], -xedges[0]]

        # Find peaks in density
        tbl = FindPeaks(im,3)
        
        if len(tbl)>0:
            # Combine clumps in the same place
            peakPositions = (yedges[tbl['x_peak']]+dy/2, -xedges[tbl['y_peak']]-dx/2)
            peaks = CombineClose(peakPositions)

            # Make clump positions the mean x and y of stars in the clump
            xcoords,ycoords,positions = MeanPeaks(newStars,peaks)

            # Track clumps through time
            clumps = TrackClumps(clumps,xcoords,ycoords,newStars)

            # Plot density plots with clumps
            #apertures = phot.CircularAperture(peaks, r=0.3)
            #norm = simple_norm(im, 'sqrt', percent=99.9)
            #plt.figure(figsize=(10,10))
            #plt.imshow(im, cmap='Greys_r', norm=norm, extent=extent)
            #apertures.plot(color='#0547f9', lw=1.5)
            #plt.title("{0} million years".format(int(start*1000)))

        start = round(start+dt,3)
    newClumps = []
    for clump in clumps:
        if clump[len(clump)-1][0]>start-5*dt:
            newClumps.append(clump)
        else:
            oldClumps.append(clump)
    clumps = newClumps
clumps = oldClumps+newClumps
for id in range(0,len(clumps)):
    clumps[id] = [id+1]+clumps[id]
print('{0} clumps total'.format(len(clumps)))
```
