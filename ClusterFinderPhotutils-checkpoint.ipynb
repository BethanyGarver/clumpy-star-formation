```python
import photutils as phot
from astropy.stats import sigma_clipped_stats
from astropy.table import Table
import numpy as np
import scipy.stats as stats
from scipy.interpolate import UnivariateSpline
import pylab as plt
from astropy.visualization import simple_norm
```

```python
def FitsRead(time):
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
# Read txyz file
txyz = Table.read('txyzform.fits',1)
rho = np.sqrt(txyz['XFORM']**2+txyz['YFORM']**2)
txyz.add_column(rho,name='RHO')
start = 1.6
dt = 0.001

# Read snapshot file
snapshot = FitsRead(16)
r = np.sqrt(snapshot['X']**2+snapshot['Y']**2)
theta = np.arctan2(snapshot['Y'],snapshot['X'])
vth = -snapshot['VX']*np.sin(theta)+snapshot['VY']*np.cos(theta)
snapshot.add_column(r,name='R')
snapshot.add_column(vth,name='VTH')

# Interpolate vDisk as a function of rho
bins = int(np.max(r)/2.5)
medians,binEdges,binNumber = stats.binned_statistic(r,vth,'median',bins)
edges = []
for i in range(0,len(binEdges)-1):
    edges.append((binEdges[i]+binEdges[i+1])/2)
w = np.isnan(medians)
medians[w] = 0
vDisk = UnivariateSpline(edges,medians,~w)

#Snapshot density plot
plt.figure(figsize=(10,10))
imS, xedgesS, yedgesS  = np.histogram2d (-snapshot['Y'],snapshot['X'],(100,100),range=[[-7,7],[-7,7]])
extentS = [yedgesS[0], yedgesS[-1], -xedgesS[-1], -xedgesS[0]]
plt.imshow(np.log(imS),extent=extentS)

#Plot a clump
inclump = ((snapshot['X']-2.77)**2+(snapshot['Y']-3.5)**2<0.01)
clump = snapshot[inclump]
plt.figure(figsize=(10,10))
imC, xedgesC, yedgesC  = np.histogram2d (-clump['Y'],clump['X'],(100,100))
extentC = [yedgesC[0], yedgesC[-1], -xedgesC[-1], -xedgesC[0]]
plt.imshow(np.log(imC),extent=extentC)

#Plot vDisk and density plot
plt.figure(figsize=(10,10))
imVth, xedgesVth, yedgesVth  = np.histogram2d (-vth,r,(100,100),range=[[-400,50],[0,20]])
extentVth = [yedgesVth[0], yedgesVth[-1], -xedgesVth[-1], -xedgesVth[0]]
plt.imshow(np.log(imVth),extent=extentVth)
x = np.arange(0,20)
plt.plot(x,vDisk(x),color='r')
plt.scatter(clump['R'],clump['VTH'])
plt.axes().set_aspect(20/450, 'datalim')

#Find stars that formed in time interval
end = start+20*dt
tRange = (txyz['TFORM']>start,txyz['TFORM']<end)
inRange = np.array(np.zeros(len(tRange[0])),dtype=bool)
for i in range(0,len(inRange)):
    if tRange[0][i] and tRange[1][i]:
        inRange[i] = True
newStars = txyz[inRange]

# Move stars back in time to beginning of interval
vcirc = vDisk(newStars['RHO'])
dtime = newStars['TFORM']-start
lindist = vcirc*dtime
nrot = lindist/(2*np.pi*newStars['RHO'])
nrotints = []
for i in nrot:
    nrotints.append(int(i))
nrotints = np.array(nrotints)
rotangle = nrot-nrotints
rotanglerad = 2*np.pi*rotangle
xrot = newStars['XFORM']*np.cos(-rotanglerad)-newStars['YFORM']*np.sin(-rotanglerad)
yrot = newStars['XFORM']*np.sin(-rotanglerad)+newStars['YFORM']*np.cos(-rotanglerad)
im, xedges, yedges  = np.histogram2d (-yrot,xrot,(100,100))
#extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
#fig = plt.figure(figsize=(10,10))
#plt.imshow(np.log(im),extent=extent)
#plt.title("{0} million years".format(int(start*1000)))
#start = round(start+dt,3)
```

```python
# Find peaks in density
data = im
mean, median, std = sigma_clipped_stats(data, sigma=3.0)
threshold = median + (5 * std)
tbl = phot.find_peaks(data, threshold, box_size=11)
tbl['peak_value'].info.format = '%.8g'
print(len(tbl))
print(tbl[:10])
```

```python
positions = (tbl['x_peak'], tbl['y_peak'])
apertures = phot.CircularAperture(positions, r=5)
norm = simple_norm(data, 'sqrt', percent=99.9)
plt.figure(figsize=(10,10))
plt.imshow(data, cmap='Greys_r', origin='lower', norm=norm)
apertures.plot(color='#0547f9', lw=1.5)
plt.xlim(0, data.shape[1]-1)
plt.ylim(0, data.shape[0]-1)
```
