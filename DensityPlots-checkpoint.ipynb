```python
from astropy.table import Table
import pylab as plt
import numpy as np
import scipy.stats as stats
```

```python
def SGDensityPlot(time):
    if time < 10:
        time = '0{0}'.format(time)
    fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
    cat = Table.read(fileName,1)
    start = 0
    for i in cat['P']:
        if i == 's':
            break
        else:
            start = start+1
    end = len(cat)-1
    im, xedges, yedges  = np.histogram2d (cat['X'][start:end],cat['Y'][start:end],(100,100))
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    plt.imshow(np.log(im),extent=extent)
```

```python
def StarsDensityPlot(time):
    if time < 100:
        time = '0{0}'.format(time)
    fileName = 'coords.chem733DiffFB10T1.0.0{0}0.fits'.format(time)
    cat = Table.read(fileName,1)
    im, xedges, yedges  = np.histogram2d (cat['X'],cat['Y'],(100,100))
    extent = [xedges[0], xedges[-1], yedges[0], yedges[-1]]
    plt.imshow(np.log(im),extent=extent)
```

```python
def MakePlot(time):
    starsList = list(np.linspace(10,20,11))+list(np.linspace(30,100,8))
    if time in starsList:
        StarsDensityPlot(time)
    else:
        SGDensityPlot(time)
```

```python
allTime = list(np.linspace(1,40,40))+list(np.linspace(50,100,6))
for i in range(0,len(allTime)):
    allTime[i] = int(allTime[i])
```

```python
plt.figure(figsize=(15,50))
for i in range(0,len(allTime)):
    if i < 40:
        plt.subplot(10,4,i+1,title="{0}00 Million Years".format(allTime[i]))
        MakePlot(allTime[i])
        continue
    elif i == 40:
        plt.figure(figsize=(15,10))
    plt.subplot(2,3,i-39,title="{0}00 Million Years".format(allTime[i]))
    MakePlot(allTime[i])
```
