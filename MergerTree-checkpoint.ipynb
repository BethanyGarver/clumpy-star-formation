```python
from astropy.table import Table
import numpy as np
import pylab as plt
```

```python
def FitsRead(time):
    # Read the fits file for that time and return table of stars
    if time < 10:
        time = '0{0}'.format(time)
    try:
        fileName = 'coords.starsandgas.chemrun733DiffFB10T1.0.00{0}0.fits'.format(time)
        read = Table.read(fileName,1)
        stars = (read['P']=='s')
        cat = read[stars]
    except:
        try:
            fileName = 'coords.chem733DiffFB10T1.0.00{0}0.fits'.format(time)
            cat = Table.read(fileName,1)
        except:
            raise ValueError('No file')
    return cat
```

```python
txy = Table.read('clumps5.fits',1)
txy.add_column(Table.Column(np.zeros(len(txy))+2),name='ID')
for i in range(0,len(txy)):
    if txy['CLUMPID'][i]==1:
        txy['ID'][i] = 1
stars = Table.read('clumps5.fits',2)
merges = Table.read('clumps5.fits',3)
merges.reverse()
txyz = Table.read('txyzform.fits',1)
for merge in merges:
    for i in txy:
        if i['CLUMPID']==merge['CLUMP1']:
            id1 = i['ID']
            break
    for i in range(0,len(txy)):
        if txy['CLUMPID'][i]==merge['CLUMP2']:
            txy['ID'][i] = id1+1
        elif txy['ID'][i]>id1:
            txy['ID'][i] = txy['ID'][i]+1
idmax = np.max(txy['ID'])
for i in range(0,len(txy)):
    if txy['ID'][i]==idmax:
        txy['ID'][i] = txy['ID'][i]+txy['CLUMPID'][i]
idlist = np.unique(np.array(txy['ID']))
for i in range(0,len(idlist)):
    for j in range(0,len(txy)):
        if txy['ID'][j]==idlist[i]:
            txy['ID'][j] = i+1
    #for j in range(0,len(merges)):
    #    for k in range(2,len(merges[j])):
    #        if merges[j][k]==idlist[i]:
    #            merges[j][k] = i+1
idlist = np.unique(np.array(txy['ID']))
for i in range(0,len(merges)):
    for j in range(2,len(merges[i])):
        if merges[i][j]!=0:
            inclump = txy[(txy['CLUMPID']==merges[i][j])]
            merges[i][j] = inclump['ID'][0]
```

```python
plt.figure(figsize=(20,20))
for i in idlist:
    clump = txy[(txy['ID']==i)]
    plt.plot(clump['TIME'],clump['ID'],c='k')
for merge in merges:
    merge = list(merge)
    ids = merge[2:2+merge[1]]
    t = [merge[0]]*merge[1]
    plt.plot(t,ids,c='k')
    plt.scatter(t,ids,c='r')
```
